# Data processing

## Import raw data

```python
import pandas as pd
import sklearn
import matplotlib

nrl = pd.read_excel('../data/raw/nrl.xlsx', header=1)

# Take a look
nrl.head()
```

```python
list(nrl.columns)
```

```python
# Drop observations before a certain year
mask = pd.to_datetime(nrl['Date']) > '2017-01-01'
nrl.query('@mask', inplace = True)

# Get columns of interest
nrl = nrl[['Date', 'Home Score', 'Away Score', 'Home Odds', 'Away Odds']]

# Derive margin variable; this is our response variable
nrl['Margin'] = nrl['Home Score'] - nrl['Away Score']
```

# Write to CSV

```python
nrl.to_csv('../data/processed/nrl.csv')
```
